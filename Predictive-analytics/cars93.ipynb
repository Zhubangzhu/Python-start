# groupby and apply

```python
def top5 (df,n=5,column = 'Price'):
    return df.sort_values(by=column)[-n:]
```

```python
df.groupby('DriveTrain').apply(top5)
```

```python
df4 = df.groupby('DriveTrain').apply(top5)
```

```python
df5 = df4[['Manufacturer','Model','Price']]
```
